```python
import numpy as np
import plotly.graph_objs as go

# create the trajectory data
trajectory = np.random.randn(320, 24, 2)

# create a list of Scatter objects for each frame
data = []
for i in range(320):
    frame = go.Scatter(
        x=trajectory[i,:,0],  # x-coordinates of particles for this frame
        y=trajectory[i,:,1],  # y-coordinates of particles for this frame
        mode='markers',
        marker=dict(
            size=5,
            color='blue'
        )
    )
    data.append(frame)

# create the layout
layout = go.Layout(
    xaxis=dict(range=[-10, 10]),
    yaxis=dict(range=[-10, 10]),
    title='Particle Trajectory Animation'
)

# create the animation
frames = [go.Frame(data=[d]) for d in data]
animation = go.Figure(frames=frames, layout=layout)

# show the animation
animation.show()
animation.frame = dict(duration=50, redraw=True)
```
